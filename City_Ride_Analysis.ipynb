```python
#Load in city csv data source
city_df = pd.read_csv("Resources/city_data.csv")
city_df.head(10)
```

```python
#Merging DataFrames
ride_city_df = pd.merge(ride_df,city_df,how="left",on=["city","city"])
ride_city_df
```

```python
#Create Urban City Data Frame
urban_cities_df = ride_city_df[ride_city_df["type"]=="Urban"]
urban_cities_df.head()
```

```python
#Create Suburban Data Frame
suburban_cities_df = ride_city_df[ride_city_df["type"]=="Suburban"]
suburban_cities_df.head()
```

```python
rural_cities_df = ride_city_df[ride_city_df["type"]=="Rural"]
rural_cities_df
```

```python
#Number of rides for urban cities
urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.head()
```

```python
#Number of rides for suburban cities
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count.head()
```

```python
#Number of rides for Rural cities
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count.head()
```

```python
#Avg fare for each city in urban cities
urban_avg_fare = urban_cities_df.groupby(["city"]).mean()["fare"]
urban_avg_fare.head()
```

```python
#Avg fare for each city in suburbs
suburban_avg_fare = suburban_cities_df.groupby(["city"]).mean()["fare"]
suburban_avg_fare.head()
```

```python
#Avg fare for each Rural City
rural_avg_fare = rural_cities_df.groupby(["city"]).mean()["fare"]
rural_avg_fare.head()
```

```python
#Avg Number of Drivers for Each Urban City
urban_driver_count = urban_cities_df.groupby(["city"]).mean()["driver_count"]
urban_driver_count.head()
```

```python
#Avg Number of Drivers for Each city in Suburbs
suburban_driver_count = suburban_cities_df.groupby(["city"]).mean()["driver_count"]
suburban_driver_count.head()
```

```python
rural_driver_count = rural_cities_df.groupby(["city"]).mean()["driver_count"]
rural_driver_count.head()
```

```python
#Build the scatter plots for urban cities
plt.scatter(urban_ride_count,urban_avg_fare, s=10*urban_driver_count,
           color="coral",edgecolor="black", linewidths=1,
           alpha=0.8,label="Urban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Avg Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
#Scatter plot for Suburbs
plt.scatter(suburban_ride_count,suburban_avg_fare,color="skyblue",
            s=10*suburban_driver_count,edgecolor="black",linewidths=1,
           alpha=0.8,label="Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
#Scatter plot for Rural Cities
plt.scatter(rural_ride_count,rural_avg_fare,color="gold",
           edgecolor="black",linewidths=1,alpha=0.8,label="Rural",
           s=10*rural_driver_count)
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Avg Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
plt.legend()
```

```python
# Build the scatter charts for each city type.
plt.subplots(figsize=(10, 6))
plt.scatter(urban_ride_count,
      urban_avg_fare,
      s=10*urban_driver_count, c="coral",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Urban")

plt.scatter(suburban_ride_count,
      suburban_avg_fare,
      s=10*suburban_driver_count, c="skyblue",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Suburban")

plt.scatter(rural_ride_count,
      rural_avg_fare,
      s=10*rural_driver_count, c="gold",
      edgecolor="black", linewidths=1,
      alpha=0.8, label="Rural")

# Incorporate the other graph properties
plt.title("PyBer Ride-Sharing Data (2019)", fontsize=20)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.xlabel("Total Number of Rides (Per City)", fontsize=12)
plt.grid(True)

#Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Add the legend.
plt.legend()

#Incorporate a text label about the circle size.
plt.text(42,35,"Note: Circle size correlates with driver count per city.",
        fontsize="12")

# Show the plot
plt.show()

#Save the figure to Analysis folder
plt.savefig("Analysis/Fig1.png")
```
